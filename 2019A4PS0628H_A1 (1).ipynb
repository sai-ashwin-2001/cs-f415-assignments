Q8) The hotel administration decided to compare the overall statistics. So for this initially they want to compare the market segment with respect to agent.Help the administration by plotting a bar graph between market segment and agent. Also find the median agents for each type of market segment.

**Task-1**: Plot a graph between market segment and agent. [1/2 mark]

**Task-2**: Find the median agents for each type of market segment. [1/2 mark]

```python
lab = sheet1.market_segment.unique()
lab = lab.tolist()
d = {}
for l in lab:
  d[l] = []
m = sheet1.shape[0]
for i in range(m):
  for l in lab:
    if l == sheet1['market_segment'][i]:
      if not np.isnan(sheet1['agent'][i]):
        d[l].append(sheet1['agent'][i])
```

```python
d
```

```python
d_sum={}
for l in lab:
  d_sum[l] = sum(d[l])
```

```python
d_sum
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize = (10, 5))
names = list(d_sum.keys())
values = list(d_sum.values())
plt.bar(names,values,color ='maroon',width = 0.4)
plt.xlabel("Market Segment")
plt.ylabel("Agent")
plt.show()
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize = (10, 5))
plt.bar(sheet1['market_segment'], sheet1['agent'],color ='maroon',width = 0.4)
plt.xlabel("Market Segment")
plt.ylabel("Agent")
plt.show()
```

```python
med = {}
import statistics
for k in lab:
  try:
    med[k] = statistics.median(d[k])
  except:
    med[k] = 0.0
```
