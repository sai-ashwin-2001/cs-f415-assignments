# Task
Design Naive Bayes Classifier from scratch for the given Dataset and find out the accuracy of your model. [10 points]

```python
data1 = sheet1.sample(frac=0.8, random_state=142)
test=sheet1.drop(data1.index)
```

```python
data1 = data1.sample(frac=1).reset_index(drop=True)
```

```python
data1
```

```python
#Prior Probabilities dictionary
def dict_to_pd(coln):
  dic = {}
  for c in cl:
    dic[c] = {}
  m = data1.shape[0]
  for i in range(m):
    for c in cl:
      if c == data1['clazz'][i]:
        if data1[coln][i] in dic[c]:
          dic[c][data1[coln][i]] += 1
        else:
          dic[c][data1[coln][i]] = 1
        #dic[c].append(sheet1[coln][i])
  #dat=pd.DataFrame.from_dict(dic,orient='index').transpose()
  return dic
```

```python
d = {}
class_prior = {}
for c in cl:
  class_prior[c] = list(data1['clazz']).count(c)
t = sum((class_prior).values())
for c in cl:
  class_prior[c] = class_prior[c]/t
for col in data1.columns:
  if not col == 'clazz':
    d[col] = dict_to_pd(col)
print(class_prior)
print(d)
for col in data1.columns:
  if not col == 'clazz':
    for c in cl:
      tot = sum((d[col][c]).values())
      for k in d[col][c].keys():
        d[col][c][k] = d[col][c][k]/tot
print(d)
```

```python
#naive bayes classifier
test = test.sample(frac=1).reset_index(drop=True)
classify = []
for i in range(test.shape[0]):
  prob = {}
  for c in cl:
    p = class_prior[c]
    for col in test.columns:
      if not col == 'clazz':
        att = test[col][i]
        #print(att)
        try:
          p *= d[col][c][att]
        except:
          p = 0
      prob[c] = p
  #print(prob)
  classify.append(max(prob, key = prob.get))
print(classify)
```

```python
y_actual = pd.Series(list(test['clazz']), name= 'Actual' )
y_pred = pd.Series(classify, name= 'Predicted' )
confmat = pd.crosstab(y_actual,y_pred)
cla = list(confmat.columns)
```

```python
confmat
```

```python
confmat = np.asarray(confmat)
class_acc = np.zeros(4)
for i in range(4):
  num = confmat[i][i]
  s =0
  for j in range(4):
    s += confmat[i][j]
  class_acc[i]= num/s
acc = np.trace(confmat)/np.sum(confmat)
for c in range(4):
  print( 'Accuracy of class '+ cla[c] +' : ' + str(class_acc[c]*100))
print('Overall Accuracy : ' + str(acc*100))
```
