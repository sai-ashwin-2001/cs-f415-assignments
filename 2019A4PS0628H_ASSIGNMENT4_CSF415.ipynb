```python
xls = pd.ExcelFile('bread.xlsx')
df1 = pd.read_excel(xls, 'Sheet2')
df1 = df1[df1['Item'] != 'NONE']
df1 = df1.sample(frac=1).reset_index(drop=True)
```

```python
transaction = {}
m = df1.shape[0]
for i in range(m):
  if df1['Transaction'][i] not in transaction:
    transaction[df1['Transaction'][i]] = {df1['Item'][i]}
  else:
    transaction[df1['Transaction'][i]].add(df1['Item'][i])
```

```python
def powerSet(items):
    """
    Power set generator: get all possible combinations of a listâ€™s elements

    Input:
        items is a list
    Output:
        returns 2**n combination lists one at a time using a generator 

    Reference: edx.org 6.00.2x Lecture 2 - Decision Trees and dynamic programming
    """

    N = len(items)
    # enumerate the 2**N possible combinations
    for i in range(2**N):
        combo = set()
        for j in range(N):
            # test bit jth of integer i
            if (i >> j) % 2 == 1:
                combo.add(items[j])
        yield combo
```

```python
powset = set()
for k in transaction:
  for i in powerSet(list(transaction[k])):
    powset.add(frozenset(i))
```

```python
powset # generate all possible subsets from each of the transactions
```

```python
lattice = dict()
for ele in powset:
  if len(ele) not in lattice:
    lattice[len(ele)] = [set(ele)]
  else:
    lattice[len(ele)].append(set(ele))
```

```python
lattice #all possible sets grouped by length of the set
```

```python
support = dict()
#c = 0
for n in lattice.keys():
  for j in lattice[n]:
    count = 0
    for k in transaction:
      if j.issubset(transaction[k]):
        count +=1
    #c += count
    support[frozenset(j)] = count/max(transaction.keys())
  for s in support:
    if support[s]<0.04:
      for l in lattice:
        for p in lattice[l]:
          if set(s).issubset(p):
            lattice[l].remove(p) # sets less than minsup and associated supersets which they are part of are pruned by apriori principle
```

```python
for s in support:
  if support[s]>0.04:
    print(s,":",support[s])
```

```python
association_rules= dict()
for l in lattice[2]:
  el = list(l)
  #print(set(el))
  d = set()
  d.add(el[0])
  association_rules[(el[0],el[1])] = support[frozenset(l)]/support[frozenset(d)]
  d.remove(el[0])
  d.add(el[1])
  association_rules[(el[1],el[0])] = support[frozenset(l)]/support[frozenset(d)]
```
